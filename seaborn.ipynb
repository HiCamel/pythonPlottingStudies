```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd



#Import software_usage.csv into a dataframe
df = pd.read_csv('software_usage.csv')

df['date'] = pd.to_datetime(df['date'])

for softwareName in df['software'].unique():
    af = df[df['software'].str.contains(softwareName)].sort_values(by='date', ascending=True).groupby('date')['software'].count().reset_index()

    # af = elasticsearch_count_executions.to_frame().reset_index()

    af.columns = ["date", "count"]

    af.head()

    ax = sns.lineplot(data=af, x='date', y='count')
    sns.set_style('whitegrid')
    sns.set_palette('pastel')
    plt.title(f"Spring 2025 - Software Executions {softwareName}", fontsize=16)
    plt.xlabel("Date", fontsize=12)
    
    #This forces whole integers only for one axis
    plt.locator_params(axis="y", integer=True, tight=True)

    plt.ylabel("Value", fontsize=12)
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.tight_layout()       # Adjust spacing to prevent label cutoff
    plt.show()



```
