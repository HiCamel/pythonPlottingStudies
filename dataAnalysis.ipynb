```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("software_usage.csv")
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.head()
```

```python
software_usage = df.groupby('software')['usage_minutes'].sum().sort_values(ascending=False)
```

```python
software_usage.head()
```

```python
execution_count = df.groupby('software')['date'].count()
execution_count.head()
```

```python
plt.title('Software execution count in past 6 months')
plt.xlabel('Software')
plt.ylabel('count')
execution_count.plot(kind='bar', color='skyblue')
```

```python
execution_counts = df.groupby(['date', 'software']).size().unstack(fill_value=0)
execution_counts.plot(kind='bar')
plt.title("Software Usage")
plt.tight_layout()
plt.show()
```

```python
# Example from dictionary
data = {
    'Year': [2020, 2021, 2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030, 2031],
    'Sales': [250, 300, 400, 450, 250, 300, 400, 450, 250, 300, 400, 450],
    'Profit': [20, 30, 50, 70, 20, 30, 50, 70, 20, 30, 50, 70]
}
df = pd.DataFrame(data)

df.plot(x='Year', y='Sales', kind='line', marker='o')
plt.title('Sales Over Years')
plt.ylabel('Sales')
plt.grid(True)
plt.show()

```

```python
ax = df.plot(x='Year', y=['Sales', 'Profit'], figsize=(16,5), kind='line', marker='o')
ax.set_ylim(0, max(df['Sales']) + 100)
plt.title('Sales and Profit')
plt.grid(True)
plt.show()
```
